```python
import pandas as pd

def limpiar_texto(texto):
    if pd.isnull(texto):
        return ""
    return (
        str(texto)
        .strip()
        .lower()
        .replace('á', 'a')
        .replace('é', 'e')
        .replace('í', 'i')
        .replace('ó', 'o')
        .replace('ú', 'u')
        .replace('ñ', 'n')
    )

def limpiar_dataset_final(df):
    # Copia para no modificar el original
    df = df.copy()

    # Limpieza de claves importantes
    if 'nombre_peaje' in df.columns:
        df['nombre_peaje'] = df['nombre_peaje'].apply(limpiar_texto)
    if 'codigo_tramo' in df.columns:
        df['codigo_tramo'] = df['codigo_tramo'].astype(str).str.strip()

    # Convertir fechas
    for campo in ['Desde', 'Hasta']:
        if campo in df.columns:
            df[campo] = pd.to_datetime(df[campo], format='%m/%d/%Y %I:%M:%S %p', errors='coerce')

    # Crear campo 'mes' desde 'Desde'
    if 'Desde' in df.columns:
        df['mes'] = df['Desde'].dt.to_period('M').astype(str)

    # Eliminar columnas duplicadas (_x, _y)
    duplicadas = [col for col in df.columns if col.endswith('_x') or col.endswith('_y')]
    df.drop(columns=duplicadas, inplace=True)

    # Asegurar que no haya nombres repetidos
    df = df.loc[:, ~df.columns.duplicated()]

    # Renombrar columnas para BigQuery
    df.columns = (
        df.columns
        .str.strip()
        .str.lower()
        .str.replace(' ', '_')
        .str.replace('á', 'a')
        .str.replace('é', 'e')
        .str.replace('í', 'i')
        .str.replace('ó', 'o')
        .str.replace('ú', 'u')
        .str.replace('ñ', 'n')
        .str.replace(r'[^\w_]', '', regex=True)
    )

    df.reset_index(drop=True, inplace=True)

    return df
```

```python
df_final_limpio = limpiar_dataset_final(df_final)
df_final_limpio.head()
```
